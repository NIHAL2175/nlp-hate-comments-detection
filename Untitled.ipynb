```python
import pandas as pd
import re
import random
from sklearn.utils import shuffle
from indicnlp.tokenize import indic_tokenize
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import numpy as np
from bpemb import BPEmb

# =========================
# Step 1: Load Dataset
# =========================
df = pd.read_csv("dataset.csv").dropna()  # 'text' and 'label' columns
df = df[['text', 'label']]
print("Original dataset shape:", df.shape)
```

```python
# =========================
# Step 2: Preprocessing
# =========================
tamil_stopwords = set([
    "ஆக", "இல்", "உங்கள்", "என்று", "ஆனால்", "அவர்", "இவர்", 
    "என்", "எது", "ஒரு", "ஆம்", "இல்லை", "நான்", "நீ", "அது", "அந்த"
])

def preprocess_tamil(text):
    text = str(text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # remove URLs
    text = re.sub(r'[^அ-ஔக-ஹa-zA-Z\s]', '', text)      # remove punctuation/numbers
    tokens = indic_tokenize.trivial_tokenize(text)
    tokens = [t for t in tokens if t not in tamil_stopwords]
    return ' '.join(tokens)

# Apply preprocessing
df['clean_text'] = df['text'].apply(preprocess_tamil)


# Show 5 examples: original vs cleaned
for i in range(2):
    print(f"Example {i+1}")
    print("Original:", df['text'].iloc[i])
    print("Cleaned :", df['clean_text'].iloc[i])
    print("-" * 60)
```

```python
df['label_enc'] = df['label'].map({ 'Non-Hate-Speech': 0, 'Hate-Speech': 1 })
hate_df = df[df['label_enc'] == 1]
non_hate_df = df[df['label_enc'] == 0]

def augment_text(text):
    words = text.split()
    if len(words) > 2:
        i, j = random.sample(range(len(words)), 2)
        words[i], words[j] = words[j], words[i]
    return ' '.join(words)

# Identify minority
if len(hate_df) > len(non_hate_df):
    minority_df = non_hate_df
    majority_count = len(hate_df)
    minority_label = 0
else:
    minority_df = hate_df
    majority_count = len(non_hate_df)
    minority_label = 1

augmented_samples = []
num_aug = majority_count - len(minority_df)
for i in range(num_aug):
    idx = i % len(minority_df)
    augmented_samples.append(augment_text(minority_df.iloc[idx]['clean_text']))

aug_df = pd.DataFrame({
    'clean_text': augmented_samples,
    'label_enc': [minority_label]*len(augmented_samples)
})

df_balanced = pd.concat([hate_df, non_hate_df, aug_df], ignore_index=True)
df_balanced = shuffle(df_balanced, random_state=42)
print("Balanced dataset counts:\n", df_balanced['label_enc'].value_counts())

```

```python
# =========================
# Step 4: TF-IDF + BPEmb Embeddings
# =========================
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df_balanced['clean_text'])

bp_ta = BPEmb(lang="ta", vs=10000, dim=300)  # Tamil BPEmb

def text_to_vector(text):
    tokens = text.split()
    if len(tokens) == 0:
        return np.zeros(bp_ta.dim)
    vectors = bp_ta.embed(text)
    if vectors.shape[0] == 0:
        return np.zeros(bp_ta.dim)
    return vectors.mean(axis=0)

# Convert BPEmb vectors for all sentences
X_bpemb = np.array([text_to_vector(t) for t in df_balanced['clean_text']])

# Combine TF-IDF + BPEmb
from scipy.sparse import hstack
X_combined = hstack([X_tfidf, X_bpemb])

y = df_balanced['label_enc']

print("TF-IDF shape :", X_tfidf.shape)
print("BPEmb shape  :", X_bpemb.shape)
print("Combined shape:", X_combined.shape)

```

```python
# Step 5: Train-Test Split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42, stratify=y
)

# =========================
# Step 6: Train Random Forest
# =========================
rf_model = RandomForestClassifier(
    n_estimators=500, 
    class_weight='balanced',
    random_state=42
)
rf_model.fit(X_train, y_train)

# =========================
# Step 7: Evaluate
# =========================
y_pred = rf_model.predict(X_test)
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=['Non-Hate-Speech','Hate-Speech']))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# =========================
# Step 8: Save Pipeline
# =========================
pipeline = {
    'tfidf': tfidf,
    'bpemb': bp_ta,
    'model': rf_model
}
joblib.dump(pipeline, "tamil_rf_tfidf_bpembbbb.pkl")
print("✅ Random Forest + TF-IDF + BPEmb pipeline saved successfully")
```

```python
# Step 9: Predict Function
# =========================
def predict_tamil(text):
    clean = preprocess_tamil(text)
    vec_tfidf = tfidf.transform([clean])
    vec_bp = text_to_vector(clean).reshape(1, -1)
    from scipy.sparse import hstack
    vec_combined = hstack([vec_tfidf, vec_bp])
    pred = rf_model.predict(vec_combined)[0]
    prob = rf_model.predict_proba(vec_combined)[0][pred]
    label = "Hate-Speech" if pred == 1 else "Non-Hate-Speech"
    return label, prob

# =========================
# Step 10: Test Examples
# =========================
examples = [
    "நீ ஒரு கெட்டவன்",
    "நான் உன்னை நேசிக்கிறேன்",
    "அவள் அழகானவள்"
]

for e in examples:
    print(f"👉 {e} → {predict_tamil(e)}")
```
